```python
import numpy as np
```

```python
def forward(P: np.ndarray, pi: np.ndarray, A: np.ndarray, obs: np.ndarray) -> float:
    N=P.shape[0]
    M=obs.shape[0]
    
    alpha=np.zeros((N,M))
    alpha[:,0]=pi*A[:, obs[0]-1]
    
    for t in range(1,M):
        for n in range(N):
            alpha[n,t] = np.dot(alpha[:,t-1] , P[:,n]) * A[n,obs[t]-1]
            
    return alpha[:,M-1].sum()     
```

```python
pi=[.2,.8]
P=np.array( [[.5,.5] , [.4, .6]])
A=np.array([ [.5,.5,.1] , [.2, .4 , .4] ])
O=np.array([1,3,3,2,3,3,3,2,1,2,1,1,3,1,3,3,3])
forward(P, pi, A, O)
```
